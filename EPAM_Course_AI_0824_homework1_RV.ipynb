## Height and weight

Create an array of length N for the height of girls (`height_f`) filled with elements linearly growing from 145 to 165 (cm). Generate an array of a normal "noise" using `bumpy.random.normal` with zero mean and $\sigma=3$. Add this noise to the previous array.

Perform similar actions for the height of boys (`height_m`) but for heigth range [145, 175].

**Note:** Please, fix `random.seed` in the first cell of this section for the reproducibility.

Using similar procedures create an arrays for weights of girls (`weight_f`: from 35 to 60 kg) and boys (`weight_f`: from 35 to 70 kg). Set the value of $\sigma=5$.

```python
np.random.seed(42)
```

```python
noise_height = np.random.normal(0, 3, N)

height_f = np.linspace(145, 165, N)
height_f += noise_height

height_m = np.linspace(145, 175, N)
height_m += noise_height
```

```python
noise_height
```

```python
height_f
```

```python
height_m
```

```python
noise_weight = np.random.normal(0, 5, N)

weight_f = np.linspace(35, 60, N)
weight_f += noise_weight

weight_m = np.linspace(35, 70, N)
weight_m += noise_weight
```

```python
noise_weight
```

```python
weight_f
```

```python
weight_m
```

### Plot of weight and height for girls and boys as functions of age

Please plot the entitled dependencies in two frames one over other. One of them should plot height for boys and girls in one frame, while the other - weight.

Do not forget provide your figures with titles, legend and labels for axes.

```python
# Plotting the data
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12))

# Plot height
ax1.plot(range(N), height_f, label='Girls', marker='o', linestyle='-', color='pink')
ax1.plot(range(N), height_m, label='Boys', marker='o', linestyle='-', color='blue')
ax1.set_title('Height of Boys and Girls')
ax1.set_xlabel('Sample Index')
ax1.set_ylabel('Height (cm)')
ax1.legend()
ax1.grid(True)

# Plot weight
ax2.plot(range(N), weight_f, label='Girls', marker='o', linestyle='-', color='pink')
ax2.plot(range(N), weight_m, label='Boys', marker='o', linestyle='-', color='blue')
ax2.set_title('Weight of Boys and Girls')
ax2.set_xlabel('Sample Index')
ax2.set_ylabel('Weight (kg)')
ax2.legend()
ax2.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

## Tables for girls and boys

Combine your data in two dataframes with columns "age", "sex", "height", and "weight" for boys and girls separately.

```python
data_f = {'age': age_f, 'sex': sex_f, 'height': height_f, 'weight': weight_f}
girls = pd.DataFrame(data_f)
girls
```

```python
data_m = {'age': age_m, 'sex': sex_m, 'height': height_m, 'weight': weight_m}
boys = pd.DataFrame(data_m)
boys
```

## General table

Create a general table both for girls and boys sorting rows by
- at first by age (in ascending order)
- by sex (in descending order: boys first)

```python
general = pd.concat([girls, boys])
general = general.sort_values(by=['age', 'sex'], ascending=[True, False])
general.reset_index(drop=True, inplace=True)
general
```
